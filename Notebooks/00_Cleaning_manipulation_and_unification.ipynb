```python
# Function that calculates missing values per column
def missing_values_table(df):
        # Total null values
        mis_val = df.isnull().sum()
        
        # Null Values Percentage
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Create a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by the percentage of null values
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Dataframe with information
        return mis_val_table_ren_columns

missing_values_table(chemical_clean)
```

### 4) Calculation of the Water Quality Index (WQI)

The Water Quality Index (WQI) is given by the following expression:

$WQI=\sum W_{i}\cdot Q_{i}$
 
 where:
 
$$W_{i}=\frac{w_{i}}{\sum \limits_{i=1}^{n} w_{i}}$$

where:
- $ W_{i}$ represents the relative weight of each measured parameter
- $ w_{i} $ is the weight of each parameter

And:
$$ Q_{i}=\frac{C_{i}x100}{S_{i}} $$

where:

- $Q_{i}$ is the quality rating (normalized parameter)
- $ C_{i}$ is the concentration of each parameter (mg / L)
- $ S_{i}$ is the standard parameter of quality given by the WHO

To obtain the normalized parameters, we have to use the standard quality values for each of them:

```python
Si=pd.DataFrame({
    "Parameter":["TDS", "HCO3", "Cl", "NO2+NO3", "F", "Ca", "Mg", "Na", "K" ],
    "Parameter Standar (Si)": [500,120,250,45,1,75,30,200,10],
    "Wi": [0.152,0.03,0.152,0.152,0.152,0.091,0.091,0.121,0.061],
            })
Si["Wi*100/Si"]= Si["Wi"]*100/Si["Parameter Standar (Si)"]
Si
```

```python
# Create a column with the Bicarbonate (HCO3) values in the mg/L units:
chemical_stations["HCO3 (mg/L)"]=chemical_stations["[HCO3] (mmol/L)"]*61
```

```python
# Create the column with the WQI values:
chemical_stations["WQI"]=(chemical_stations["TDS (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][0])+
                 chemical_stations["HCO3 (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][1])+
                 chemical_stations["Cl (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][2])
                 +chemical_stations["NO2+NO3 (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][3])+
                chemical_stations["F (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][4])+
                 chemical_stations["Ca (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][5])+
                    chemical_stations["Mg (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][6])+
                    chemical_stations["Na (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][7])+
                   chemical_stations["K (mg/L)"].apply(lambda x: x*Si["Wi*100/Si"][8]))
```

```python
# Create the WQI classes:
col="WQI"
conditions=[(chemical_stations["WQI"]<=50), (chemical_stations["WQI"]>50) & (chemical_stations["WQI"]<=100), 
            (chemical_stations["WQI"]>100) & (chemical_stations["WQI"]<=200),
           (chemical_stations["WQI"]>200) & (chemical_stations["WQI"]<=300),chemical_stations["WQI"]>300]  
choices=["Excellent", "Good", "Poor", "Very Poor", "Unsuitable"]
chemical_stations["Class WQI"]=np.select(conditions, choices, default=np.nan)
```

### 5) Inorganic Chemistry Index (ICI) calculation

We are going to create a new column with the Inorganic Chemistry of Water Index. This index is given by the expression:

$ICI =\frac{[Cl]+2[SO4]+[NO3]+3[PO4]}{[Cl]+2[SO4]+[NO3]+3[PO4]+[HCO3]}x100$

```python
# Create the ICI for each sample:
chemical_stations["ICI (%)"]=chemical_stations.apply(lambda row: (row["[Cl] (mmol/L)"]+ 
                                       2*row["[SO4] (mmol/L)"]+ 
                                       row["[NO3] (mmol/L)"]+
                                       3*row["[PO4] (mmol/L)"])*100
                                         /(row["[Cl] (mmol/L)"]+ 
                                       2*row["[SO4] (mmol/L)"]+ 
                                       row["[NO3] (mmol/L)"]+
                                       3*row["[PO4] (mmol/L)"]+
                                        row["[HCO3] (mmol/L)"]),axis=1)
```

```python
# Create the ICI classes:

col="ICI (%)"
conditions=[chemical_stations["ICI (%)"]>70,
            (chemical_stations["ICI (%)"]>30) & (chemical_stations["ICI (%)"]<70),
            chemical_stations[col]<30]  
choices=["Non-weathering dominated", "Mix","Rock dominance"]
chemical_stations["Class ICI"]=np.select(conditions, choices, default=np.nan)
```
