```python
# Rename the first column of the dataset "dams_lakes2" so that you can open all 5 datasets at once and concatenate them:
dams_lakes2.rename({'SAMPLE NUMBER AND SAMPLING DATE': "SAMPLE STATION/POINT ID\nAND SAMPLING DATE"}, axis=1, inplace=True)
# Save this change in a new file:
dams_lakes2.to_excel('Dams_and_lakes_1999_2012.xlsx', index=False)
```

```python
# Open all 5 datasets at once to directly form a single file
import os
path=os.getcwd()
files=os.listdir(path)
files=('Dams and lakes up to 1998 (A-X).xlsx', 
       'Rivers 1999-2012 (A-X).xlsx', 
       'Rivers up to 1998 (A-D).xlsx', 
       'Rivers up to 1998 (E-X).xlsx',
       'Dams_and_lakes_1999_2012.xlsx')
# Create empty dataframe
chemical=pd.DataFrame()
# Read and join files
for f in files:
    data = pd.read_excel(f, skiprows=[1,2])  # Delete the first two rows, they are not useful
    chemical= chemical.append(data,ignore_index=True)
# Take a look at the dataframe
chemical.head()
```
